## Purpose

The current issue is that both users and clinicians would like the ability to adjust the speed and torque output of the device to have a more natural response during different use cases.  Currently the device is unable to adjust the torque and speed outputs independently of one another.  The practical effects of this are that users are not able to operate the device in a slower controlled fashion but yet have the ability to have the same high-level lifting capability as if it was operating at a higher speed.

## Variables

```python
offset = 0.0  #Deadband offset across the X- Axis

t_pos = np.linspace(offset, 1, 1000) #Domain of EMG values 

max_pace = 6 #max pace value

b = np.linspace(.1, 2, 10) #Sensitivity Exponential 
```

Max Pace:  The highest amount of delta position (degrees for the elbow and % ROM for the Grasp) that the SP can be changed within one cycle

Sensitivity:  Mathematically this is the slope but practically this is how much the input changes the output per step

Torque:  This sets the current limit that is supplied to the motors during operation

Offset:  This parameter sets the dead band (change in input no change in output) to account for the baseline or resting level of EMG input

## Max EMG Implementation
Implements the decay of an EMG signal according to the desired Decay Time,
expressed in mSecs.  The decay is exponention, and it is based on the RC
Time-Constant for RC circuits.  The Decay Time can be changed on-the-fly
by calling the proper API-function.  It can handle signal-rise too,
but we don't use it.  You wouldn't believe the amount of development
it took to get this right.  The math is a bit complex and misleading.
Time-Constant formula for an RC circuit is based on Euler's Number:


	y = 1.0 - e^-1

The algorithm implemented here lets the input signal rise instantaneously,
tracking it as it grows.  This becomes the maximum signal.  Then, when the
signal drops off below the max, it uses the decay-rate and the TC to decay
the maximum over the desired time.

Although the TC formula is simple enough, its explanation and various
implementations are typically restricted to the use of only 5 data values.
However, our need is a programmable decay-rate of X mSeconds, yet we must
still have the proper TC values at the 5 proper indices that correspond
to the decay-rate divided by 5.

Whether we have 10 samples between time-constants, or 100 samples,
we still need the 5 values at the proper key time-constant indices
to be correct.  Therefore, given a sample rate of 10, and a decay rate
of 1000, the 5 key TC values will be located at index 20, 40, 60, 80,
and 100 from the beginning of the decay, and they must still be correct.
All the samples in between those 5 key values must also be correct.
Using 1.0 as the starting charge, the 5 key values must be:
         Normal Response  -----  Inverted Response
	Decay	Rise		Decay	Rise
    TC 0	1.0000	0.0000		0.9932	0.0067
    TC 1	0.3679	0.6321		0.9816	0.0183
	TC 2	0.1353	0.8646		0.9502	0.0498
	TC 3	0.0498	0.9502		0.8646	0.1353
	TC 4	0.0183	0.9816		0.6321	0.3679
	TC 5	0.0067	0.9932		0.0000	1.0000

 This implemention properly accommodates the math for any desired sample
 rate, decay rate, and all the values in between the 5 key TC values.
 The magnitude of the starting charge, i.e. the EMG value, is irrelevant.

 This implementation will also compute the inverted response, which is more
 appropriate for our applicaiton.

```python
def _getDecayValue(invert,num_samples,count,tc_samples,charge):
    tcFactor = 0
    holder = 0
    if (invert == TRUE):
        tcFactor = 1.0 - math.exp(-(num_samples-count)/tc_samples)
        holder = (charge * tcFactor)
        
    else:
        tcFactor = 1.0 - math.exp((-count/tc_samples))
        holder = (charge - (charge * tcFactor))
    return(holder)
                                
def _getRiseValue(invert,num_samples,count,tc_samples,charge):
    tcFactor=0
    if (invert):
        tcFactor = 1.0 - math.exp(-(num_samples - count)/tc_samples)
        return(charge - (charge * tcFactor))
    else:
        tcFactor = 1.0 - math.exp(-(count/tc_samples))
        return(charge * tcFactor)



                                
def _doDecay(emgMax,charge,count,emgVal):
    decayVal;

## The easy case of a growing emgVal setting a new max and charge.
    if (emgVal >= emgMax):
        emgMax = charge = emgVal
        count = 0
    ++count

    if (emgVal < 0): 
        emgVal = 0.0

    decayVal = _getDecayValue(invert,num_samples,count,tc_samples,charge);

    if (decayVal < 0.001):  ## chop it to zero
        decayVal = 0

    if decayVal> 0:
        emgMax = decayVal
    else:
        emgMax = 0
    
    
                                
                                
```

## Algorithm Definition And Implementation

```python
max_pace = 6

offset = 0.1

b = 2

InitPos = 90

max_pos = 135
min_pos = 0

t_sample = (np.linspace(0, 1, 500))

new_pos = [1]* len(t)
new_pos[0] = InitPos

modifier = [1]* len(t)
mod = [1]* len(t)


### Function Definition
def LimitCheck(min_val,max_val,input_val):
    if input_val > max_val:
        input_val = max_val
    if input_val < min_val:
        input_val = min_val
    return input_val



def ContiousPaceFunction(amp,exp,pace_input,new_pos,min_pos,max_pos):
    for i in range(1,len(pace_input)):
        modifier[i] = (amp * ((pace_input[i]) ** exp))
        if pace_input[i] < 0:
            new_pos[i] = new_pos[i-1] - modifier[i]
        elif pace_input[i] > 0:
            new_pos[i] = new_pos[i-1] + modifier[i]    
        new_pos[i] = LimitCheck(min_pos,max_pos,new_pos[i])
        print("Index:",i,"New Pos:",new_pos[i],"Pace:", pace_input[i],"Modifier:",modifier[i],"Previous Pos:",new_pos[i-1])
    return (new_pos)
# y = ContiousPaceFunction(max_pace,b,t_sample,new_pos,min_pos,max_pos)

# plt.figure(figsize=(10,10))
# plt.plot(t,y, 'b')
# plt.xlabel('Time [Sec]')
# plt.ylabel('PAce [%]')
# plt.grid(color='k', linestyle='-', linewidth=2)
# plt.title("Simulated Pace Signal")
```
