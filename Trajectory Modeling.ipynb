```python
mu = 0
variance = 1
t = np.arange(0, 3.14, 0.1) #Domain of EMG values 
x = [-1,0,1,2,3,4,5]
y = [ 0,5,5,5,5,5,5]

amplitude   = 2*np.sin(t)
sigma = math.sqrt(variance)
# x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.plot(t, amplitude)
plt.legend(['Desired Position','Desired Velocity'])
plt.xlabel('Time [s]')
plt.ylabel('Joint Position [% ROM]')
```

```python
t = np.linspace(0, 1, 100) #Domain of EMG values 

q0 = 0
qf = 1
q_d = (qf-q0)

q = q0 + 3*q_d*(np.power(t,2)) - 2*q_d*(np.power(t,3))
v = 6*q_d*t - 6*q_d*(np.power(t,2))


plt.figure(figsize=(10,10))
plt.plot(t,q)
plt.xlabel('Time [s]')
plt.ylabel('Joint Position [Deg]')
plt.title('Position Trajectory')

plt.figure(figsize=(10,10))
plt.plot(t,v)
plt.xlabel('Time [s]')
plt.ylabel('Joint Velocity [Deg/sec]')
plt.title('Velocity Trajectory')
```
